This notebook illustrates the footprint currently available in regressis (see `regressis/footprint.py`). They were designed to give healpix mask. The base masks are saved in `regressis/data/` in NESTED scheme at `Nside=256` (ie) the accuracy is about the resolution at nside=256.

```python
from regressis import DR9Footprint, DESIFootprint

from regressis.plot import plot_moll
```

## 1. DR9Footprint:

see the documentation (at read the doc) for the explanation of all the attributs.

```python
# Load DR9 footprint:
dr9_footprint = DR9Footprint(256, mask_lmc=False, clear_south=True, mask_around_des=True, cut_desi=False)
```

```python
print("default regions: ", dr9_footprint.default_regions)
print("available regions: ", dr9_footprint.available_regions)
```

```python
for region in dr9_footprint.default_regions:
    plot_moll(dr9_footprint(region), galactic_plane=True, label=region)
```


## 2. DESIFootprint:

see the documentation (at read the doc) for the explanation of all the attributs.

```python
# Load DESI footprint:
desi_footprint = DESIFootprint(nside=256)
```

```python
print("available region: ", desi_footprint.available_regions)
print("available program: ", desi_footprint.available_programs)
```

```python
for survey in ['DA02']:
    for program in ['dark', 'bright']:
        plot_moll(desi_footprint(survey, program), galactic_plane=True, label=survey + ' - ' + program)
```
